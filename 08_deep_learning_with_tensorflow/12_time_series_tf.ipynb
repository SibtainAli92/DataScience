# Modeling

```python
df.shape
```

```python
# split the data into train, validation and test i have records make 70% for traning, 20% for validation and 10% for testing
# 70% of 35045 is 24531
# 20% of 35045 is 7009
# 10% of 35095 is 3505 
# total 35095
# train data 
train_df = df[:24531]
# validation data
val_df = df[24531:31540]
# test data 
test_df = df[31540:]

# let's check the shape of the data
train_df.shape, val_df.shape, test_df.shape
```

```python
# lets plot the data
train_df.plot(subplots=True)

# lets plot the data
val_df.plot(subplots=True)

# lets plot the data
test_df.plot(subplots=True)
```

```python
# lets normalize the data
# we will use the mean and standard deviation of the training data to normalize the data
train_mean = train_df.mean()
train_std = train_df.std()

# lets normalize the training data
train_df = (train_df - train_mean) / train_std
# lets normalize the validation data
val_df = (val_df - train_mean) / train_std
# lets normalize the test data
test_df = (test_df - train_mean) / train_std
```

```python
# lets plot the data
train_df.plot(subplots=True)

# lets plot the data
val_df.plot(subplots=True)

# lets plot the data
test_df.plot(subplots=True)
```
