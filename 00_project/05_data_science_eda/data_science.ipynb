# Data Science Salary - Exploratory Data Analysis (EDA)

## Introduction
Welcome to the "Data Science Salary - Exploratory Data Analysis (EDA)" notebook! In this notebook, we will explore a dataset containing information about data science salaries. Understanding salary trends and factors influencing salaries is crucial for both job seekers and employers in the data science field.

Exploratory Data Analysis (EDA) is an essential step in the data analysis process. Through EDA, we can gain insights into the characteristics and patterns present in the dataset, identify potential trends, outliers, and relationships between variables, and inform subsequent steps in the analysis.


<h2 style="font-family:Comic Sans MS; color:#a6a4a1">4. Distribution Analysis</h2>

```python
numerical_features = ['salary']
plt.figure(figsize=(10, 6))
for feature in numerical_features:
    sns.histplot(df[feature], kde=True, bins=20)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
plt.figure(figsize=(12, 8))
for feature in numerical_features:
    plt.subplot(2, len(numerical_features)//2, numerical_features.index(feature)+1)
    sns.histplot(df[feature], kde=True, bins=20)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Box plots for numerical features by categorical features
plt.figure(figsize=(12, 8))
for feature in numerical_features:
    plt.subplot(2, len(numerical_features)//2, numerical_features.index(feature)+1)
    sns.boxplot(data=df, x='employment_type', y=feature)
    plt.title(f'{feature} Distribution by Employment Type')
    plt.xlabel('employment_type')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

<h2 style="font-family:Comic Sans MS; color:#a6a4a1">5. Correlation Analysis</h2>

```python
plt.figure(figsize=(8, 6))
columns_of_interest = ['salary_in_usd', 'work_year', 'remote_ratio', 'salary']
df_corr = df[columns_of_interest]
correlation_matrix = df_corr.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

 
 <h2 style="font-family:Comic Sans MS; color:#a6a4a1">6. Company size analysis with Pie Char</h2>

```python
company_size_counts = df['company_size'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(company_size_counts, labels=company_size_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Company Size Distribution')
plt.axis('equal')
plt.show()
```

<h2 style="font-family:Comic Sans MS; color:#a6a4a1">7. Job Title Analysis</h2>

```python
plt.figure(figsize=(10, 6))

job_title_counts = df['job_title'].value_counts()

top_5_job_titles = job_title_counts.head(5)

sns.barplot(x=top_5_job_titles.index, y=top_5_job_titles.values)
plt.title('Top 5 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<h2 style="font-family:Comic Sans MS; color:#a6a4a1">8. Employment Type Analysis</h2>

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='employment_type')
plt.title('Employment Type Distribution')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```


<h2 style="font-family:Comic Sans MS; color:#a6a4a1">9. Salary Analysis</h2>

```python
plt.figure(figsize=(10, 6))

df_sorted = df.sort_values(by='salary', ascending=False)

top_5_salaries = df_sorted.head(5)

sns.barplot(x='job_title', y='salary', data=top_5_salaries)
plt.title('Top 5 Highest Salaries by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
